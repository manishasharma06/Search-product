```python
import pymongo


# establish a connection to localhost
# Connect to the MongoDB client
client = pymongo.MongoClient("mongodb://localhost:27017/")

#check whether the connection is established
if client:
    try:
         #ping the server
         client.admin.command('ping')
         print('Connection established.')
    except: 
         print('Connection failed.')
else:
    print('Unable to connect to MongoDB.')
```

```python
# select a database
db = client['Shoppingsystem']

# select a collection
collection = db['Productlist']
```

```python
# Define the search filters
#category = 'Creams','Household'
#min_rating = 4.4,3.6
#max_price = 220,800
category = input("Enter category: ")
min_rating = float(input("Enter minimum rating: "))
max_price = float(input("Enter maximum price: "))
```

```python
# Search for products that match the filters
query = {"Category": category, "Rating": {"$gte": min_rating}, "Price": {"$lte": max_price}}
results = collection.find(query)
```

```python
# Print the matching products
for product in results:
    print(product)
```

```python
cart = db['Cart']

product_name = input("Enter product name: ")
#product_name = "Falero Wah Wah Candy"
```

```python
# Find the product in the product list
product = collection.find_one({"Product_Name": product_name})
```

```python
# Check if the product already exists in the cart
cart_item = cart.find_one({"Product_Name": product_name})

# If the product already exists in the cart, increment the quantity
qunatity = int(input("Enter qunatity: "))
if cart_item:
    cart.update_one({"Product_Name": product_name}, {"$inc": {"Quantity": qunatity}})
    print("Product quantity updated in cart.")
# If the product does not exist in the cart, add it with a quantity 
else:
    cart.insert_one({
        "Product_Name": product["Product_Name"],
        "Category": product["Category"],
        "Price": product["Price"],
        "Quantity": qunatity
    })
    print("Product added to cart.")
```

```python
#filter = {"Product_Name": "Falero Wah Wah Candy"}
item_name = input("Enter the item name to delete: ")

# Define the filter to find the item to remove from the Cart collection
filter = {"Product_Name": item_name}


# Delete the item from the Cart collection
result = cart.delete_one(filter)

# Print the number of documents deleted
print("Number of items deleted:", result.deleted_count)
```

```python
# Get the item name and quantity from the user
item_name = input("Enter the item name to delete: ")
quantity = int(input("Enter the quantity to delete: "))

# Define the filter to find the item to remove from the Cart collection
filter = {"Product_Name": item_name, "Quantity": {"$gte": quantity}}

# Find the item in the Cart collection and remove the specified quantity
result = cart.update_one(filter, {"$inc": {"Quantity": -quantity}})

# Check if the item was found and updated
if result.matched_count == 1:
    print("Deleted", quantity, "items of", item_name)
else:
    print("Item not found in Cart collection or quantity is insufficient.")
```
